### - Pre-Assesing

 #### - Firstly identiftying the top 10 countries with the highest average growth over the 1st decade of the 21st century.

```python
top10_contrs=list(df_growth.set_index('country').loc[:,'2000':'2009'].mean(axis=1).sort_values(ascending= False).head(10).index)
```

 #### - Secondly trim/adjust all data frames as per the top10 countries list "top10_contrs".

```python
dfs=[df_gdp, df_income, df_edu_spend, df_hlth_spend, df_life_expct, df_co2, df_hdi, df_growth]

i= 0
for df in dfs:
    # Filter data frame as per the top10 countries
    df= df.set_index('country').loc[:,'2000':'2009']
    df= df[df.index.isin (top10_contrs)]
    
    #Set countries in a Descending sort.
    tmp= df_growth.set_index('country').loc[:,'2000':'2009'].mean(axis=1).sort_values(ascending= False).head(10).reset_index()
    df= df.reindex(index= tmp['country'])
    df= df.reset_index()
    
    # Modifying "dfs" list elements
    dfs[i]= df
    i+=1
    
# Assiging list elements to global data frames
df_gdp= dfs[0]
df_income= dfs[1]
df_edu_spend= dfs[2]
df_hlth_spend= dfs[3]
df_life_expct= dfs[4]
df_co2= dfs[5]
df_hdi= dfs[6]
df_growth= dfs[7]
```

 #### - Finally change the layout of loaded data frames (transpose) and combine all together in one master data frame.

```python
# Define lists of data frames, extension for each column name rpresenting each metric, and country abbreviations.
dfs=[df_gdp, df_income, df_edu_spend, df_hlth_spend, df_life_expct, df_co2, df_hdi, df_growth]
ext= ['_gdp', '_income', '_edu_spend', '_hlth_spend', '_life_expct', '_co2', '_hdi', '_growth']
con_abrv= ['GNQ','AZE','QAT','AGO','UAE','LBN','SLE','AFG','GAB','ARM']

# Set the "country" column to be the index of each data framethen transpose it.
# Finally rename each country column to be a combination of country abbreviation and each metric extension.
i=0
for df in dfs:
    df= df.set_index('country')
    df= df.transpose()
    j=0
    for col in df.columns:
        df.rename(columns= {col : con_abrv[j]+ext[i]}, inplace=True)
        j+=1
    dfs[i]= df
    i+=1
    
# Assiging modified list elements to the global data frames
df_gdp= dfs[0]
df_income= dfs[1]
df_edu_spend= dfs[2]
df_hlth_spend= dfs[3]
df_life_expct= dfs[4]
df_co2= dfs[5]
df_hdi= dfs[6]
df_growth= dfs[7]

# Merge all data frames into a single master data frame.
df_master=pd.concat([df_growth, df_gdp,df_co2, df_income, df_edu_spend, df_hdi, df_hlth_spend, df_life_expct], axis=1)
```

##### - Define
`AFG_edu_spend`

No available data for this metric for Afghanistan. 

But, I'll assume that Afghanistan keeps the same education expenditure percentage of its GDP over the years using **2009** value. 

Equation will be like `2008_edu_spend= 2009_edu_spend / ( 1 + 2009_growth/100 )`

##### Code

```python
years= [2008,2007,2006,2005,2004,2003,2002,2001,2000]

for y in years:
    df_master_clean.loc[str(y),'AFG_edu_spend']= \
                          df_master_clean.loc[str(y+1), 'AFG_edu_spend'] / (1 + df_master_clean.loc[str(y+1), 'AFG_growth']/100)
```

`LBN_hdi`

As there is no data related to this metric for Lebanon, I'll consider the shared data for the Ararb States within UNDP reports.

##### Code

```python
years= ['2000','2001','2002','2003','2004']
missing_hdi= pd.read_csv('missing_hdi.csv')

for y in years:
    df_master_clean.loc[y, 'LBN_hdi']= float(missing_hdi.loc[ missing_hdi['group'] == 'Arab States',y ])
```
